```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.mixture import GaussianMixture
```

```python
X_scaled = pd.read_csv("../datasets/preprocessed/Preprocessed_data_1.csv")
X_scaled.head()
```

```python
X_scaled.drop("Unnamed: 0",axis=1)
```

```python
numpy_X = X_scaled.to_numpy()
print(numpy_X)
```

```python
n_components = range(1, 11)
models = [GaussianMixture(n, covariance_type='full', random_state=42).fit(numpy_X)
          for n in n_components]

plt.figure(figsize=(10, 5))
plt.plot(n_components, [m.bic(numpy_X) for m in models], label='BIC', marker='o')
plt.plot(n_components, [m.aic(numpy_X) for m in models], label='AIC', marker='o')
plt.legend(loc='best')
plt.xlabel('Number of Clusters (n_components)')
plt.ylabel('Score')
plt.title('GMM Model Selection: BIC & AIC')
plt.show()
```

```python
optimal_n = 4
gmm = GaussianMixture(n_components=optimal_n, covariance_type='full', random_state=42)
gmm.fit(X_scaled)
```

```python
df = pd.read_csv('orginal_filtered_data.csv')
```

```python
# hard levels
df['cluster'] = gmm.predict(X_scaled)
```

```python
print(df['cluster'])
```

```python
df['cluster'].value_counts()
```

```python
probs = gmm.predict_proba(X_scaled)
```

```python
df['probs_clusters_0'] = probs[:,0]
df['probs_clusters_1'] = probs[:,1]
df['probs_clusters_2'] = probs[:,2]
df['probs_clusters_3'] = probs[:,3]
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(numpy_X)
```

```python
pca_df = pd.DataFrame(data=X_pca, columns=['PC1','PC2'])
pca_df['Clusters'] = df['cluster'].values
plt.figure(figsize=(10,7))
sns.scatterplot(x='PC1',y='PC2',hue='Clusters',data=pca_df, palette='tab10',alpha=0.5)
plt.title('GMM Custers Visulized via pca')
plt.show()          
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, calinski_harabasz_score
sil_score = silhouette_score(X_scaled, df['cluster'])
db_index = davies_bouldin_score(X_scaled, df['cluster'])
ch_index = calinski_harabasz_score(X_scaled, df['cluster'])

print(f"Silhouette Score: {sil_score:.4f}")
print(f"Davies-Bouldin Index: {db_index:.4f}")
print(f"Calinski-Harabasz Index: {ch_index:.4f}")
```

```python
analysis = df.groupby('cluster')[['Quantity', 'UnitPrice', 'Hours', 'DayOfweek']].mean()
print(analysis)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
pipeline = Pipeline([
    ('scaler',StandardScaler()),
    ('gmm',GaussianMixture(n_components=4, covariance_type='full', random_state=42))
])
pipeline.fit(df[['Quantity', 'UnitPrice', 'Hours', 'DayOfweek']])
```
